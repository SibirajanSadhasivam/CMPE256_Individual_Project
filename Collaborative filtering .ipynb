```python
# Compute the Cosine Similarity matrix based on the count_matrix
from sklearn.metrics.pairwise import linear_kernel
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(df1,df1)
```

```python
indices = pd.Series(df.index, index=df['UserID'])
```

```python
# Function that takes in company name as input and outputs most similar companies
def get_recommendations(UserID, cosine_sim=cosine_sim):
    # Get the index of the company that matches the title
    idx = indices[UserID]

    # Get the pairwsie similarity scores of all company with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the company based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    #sim_scores = sort(sim_scores, reverse=True)


    # Get the scores of the 10 most similar ones
    sim_scores = sim_scores[1:2]

    # Get the indices
    company_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return df['UserID'].iloc[company_indices]
```

```python
result=get_recommendations("Sri", cosine_sim)
r=result.values
s=r[0]
rows=df.loc[df['UserID'] == s]
```
